# Data Preprocessing

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
numeric_columns = dev_data.select_dtypes(include=['float64' , 'int64']).columns
exclude_columns = ['account_number', 'bad_flag']
numeric_columns = [col for col in numeric_columns if col not in exclude_columns]

non_null_columns = dev_data[numeric_columns].columns[dev_data[numeric_columns].notna().any()]
dev_data_non_null = dev_data[non_null_columns]
```

```python
dev_data_non_null.head()
```

```python
dev_data_non_null.shape
```

```python
dev_data_non_null.isnull().sum()
```

```python
# Drop columns which contain the NaN value at all

imputer = SimpleImputer(strategy='mean')
imputed_data = imputer.fit_transform(dev_data_non_null)
dev_data_imputed = pd.DataFrame(imputed_data, columns=dev_data_non_null.columns)

# Sclae the numeric feature
scaler = StandardScaler()
scaled_data = scaler.fit_transform(dev_data_imputed)
dev_data_scaled = pd.DataFrame(scaled_data, columns=dev_data_imputed.columns)

# Merge back with non-scaled columns
dev_data_processed = dev_data.copy()
dev_data_processed.update(dev_data_scaled)

```

```python
dev_data_processed.head()
```

```python
dev_data_processed.isnull().sum()
```

# Feature Selection


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
```

```python
# Try to store the feature and target column in X amd y
X = dev_data_processed.drop(['bad_flag'] , axis=1)
y = dev_data_processed['bad_flag']


# Define the algorithms
rf = RandomForestClassifier(random_state=42)
rf.fit(X , y)
```

```python
X.head()
```

```python
# Calculate the feature importance
feature_importance = pd.Series(rf.feature_importances_ , index=X.columns)
selected_features = feature_importance[feature_importance > 0.0008230452674897119].index.tolist() # Same feature also select in the case validation data
X_selected = X[selected_features]
```

# Model Training

```python
# Import library
from sklearn.model_selection import train_test_split , cross_val_score
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
```

```python
from sklearn.metrics import roc_auc_score , roc_curve
```

```python
# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)
```

```python
# Train a Gradient Boosting model
model = GradientBoostingClassifier(random_state=42)
model.fit(X_train, y_train)
```

# Model Evaluation

```python
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[: , 1]

# Calculate the auc
auc = roc_auc_score(y_test , y_pred_proba)
print("AUC:->",auc)
print(classification_report(y_test , y_pred))
```

```python
accuracy = accuracy_score(y_test , y_pred)
print(accuracy * 100)
```
